```python
import xml.etree.ElementTree as ET
import pandas as pd

xml_data = open('dirty_dataset.txt', 'r').read()  # Read file
root = ET.XML(xml_data)  # Parse XML

data = []
cols = []
for i, child in enumerate(root):
    data.append([subchild.text for subchild in child])
    cols.append(child.tag)

dirty_data = pd.DataFrame(data).T  # Write in dirty_data and transpose it
dirty_data.columns = cols  # Update column names
dirty_data.head(20)
```

```python
dirty_data = dirty_data.iloc[3]
```

```python
dirty_data = dirty_data.reset_index(0)
```

```python
dirty_data = dirty_data.drop('index', axis=1)
```

```python
dirty_data = dirty_data.rename(columns={3: "Verse"})
```

```python
dirty_data = dirty_data['Verse'].str.replace('\n', '_')
```

```python
dirty_data['Verse'] = dirty_data['Verse'].str.lower()
```

```python
dirty_data.to_csv('cleaned_dataset.csv')
```
